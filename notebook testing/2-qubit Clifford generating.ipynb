```python
from qiskit.quantum_info import random_clifford, Clifford
from qiskit import QuantumCircuit
```

```python
import pickle

def save_object(obj, filename):
    with open(filename, 'wb') as outp:  # Overwrites any existing file.
        pickle.dump(obj, outp, pickle.HIGHEST_PROTOCOL)
```

# STAGE 1
Obtain 11516 distinct Cliffords

```python
generated_cliffords = [random_clifford(2) for i in range(150000)]
clifford_group = []

for clifford in generated_cliffords:
    if not clifford in clifford_group:
        clifford_group.append(clifford)
        if len(clifford_group) == 11520:
            break
        
print(len(clifford_group))
```

```python
print(len(clifford_group))
print(type(clifford_group[0]))
```

```python
save_object(clifford_group, 'clifford_2qubits.pkl')
```

```python
with open('clifford_2qubits.pkl', 'rb') as inp:
    restored_cliffords = pickle.load(inp)
```

```python
generated_cliffords = [random_clifford(1) for i in range(1000)]
clifford_group = []

for clifford in generated_cliffords:
    if not clifford in clifford_group:
        clifford_group.append(clifford)
        if len(clifford_group) == 24:
            break
        
print(len(clifford_group))
```

```python
save_object(clifford_group, 'clifford_1qubits.pkl')
```
