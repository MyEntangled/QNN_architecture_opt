# 1. Euclidean, single-variate, single-valued optimization

```python
from dragonfly import minimise_function, maximise_function
import dragonfly
```

```python
func = lambda x: x ** 4 - x**2 + 0.1 * x
domain = [[-10, 10]]
max_capital = 100
```

```python
min_val, min_pt, history = minimise_function(func, domain, max_capital)
print(min_val, min_pt)
```

```python
max_val, max_pt, history = maximise_function(lambda x: -func(x), domain, max_capital)
print(max_val, max_pt)
```

```python
max_val, max_pt, history = maximise_function(lambda x: func(x), domain, max_capital)
print(max_val, max_pt)
```

# 2. Euclidean, multivariate, multi-valued optimization

```python
from dragonfly import multiobjective_maximise_functions
```

```python
def func_1(var):
    x,y = var
    z = x**3 + y**3 + 3*x*y
    return z
    
def func_2(var):
    x,y = var
    t = -3*x**2 + 3*y
    return t
    
domain = [[-10, 10],[-10, 10]]
max_capital = 100
```

```python
pareto_max_values, pareto_points, history = multiobjective_maximise_functions([func_1,func_2], domain, max_capital)
```

# 3. Discrete Euclidean, multi-variate optimization

```python
import numpy as np
from dragonfly import load_config, maximise_function
```

```python
def objective(x):
  """ Objective. """
  return np.linalg.norm(x[0])

size = 1000
dim = 3
disc_euc_items = list(np.random.random((size, dim)))
domain_vars = [
{'type': 'discrete_euclidean', 'items': disc_euc_items},
]
config_params = {'domain': domain_vars}
config = load_config(config_params)
max_num_evals = 100
```

```python
domain_vars
```

```python
# specify optimisation method
opt_method = 'bo' # Bayesian optimisation
# opt_method = 'ea' # evolutionary algorithm
# opt_method = 'rand' # random search
opt_val, opt_pt, history = maximise_function(objective, config.domain, max_num_evals,
                                           config=config, opt_method=opt_method)
print(opt_pt, opt_val)
```

# 4. Mixed domain (still Euclidean) optimization

```python
def hartmann6_3(x):
    """ Hartmann function in 3D. """
    #print(x.shape)
    print(x)
    pt = np.array([x[1][1]/11.0,
             x[0][0],
             x[0][2],
             x[0][1],
             x[2]/114.0,
             x[1][0]/11.0,
            ])
    A = np.array([[  10,   3,   17, 3.5, 1.7,  8],
            [0.05,  10,   17, 0.1,   8, 14],
            [   3, 3.5,  1.7,  10,  17,  8],
            [  17,   8, 0.05,  10, 0.1, 14]], dtype=np.float64)
    P = 1e-4 * np.array([[1312, 1696, 5569,  124, 8283, 5886],
                   [2329, 4135, 8307, 3736, 1004, 9991],
                   [2348, 1451, 3522, 2883, 3047, 6650],
                   [4047, 8828, 8732, 5743, 1091,  381]], dtype=np.float64)
    log_sum_terms = (A * (P - pt)**2).sum(axis=1)
    alpha = np.array([1.0, 1.2, 3.0, 3.2])
    return alpha.dot(np.exp(-log_sum_terms))
```

```python
## QUICK TEST
disc_euc_items = list(np.random.random((1000, 3)))
domain_vars = [{'type': 'discrete_euclidean', 'items': disc_euc_items},
             {'type': 'float',  'min': 0, 'max': 11, 'dim': 2},
             {'type': 'int',  'min': 0, 'max': 114},
            ]
config_params = {'domain': domain_vars}
config = load_config(config_params)
max_num_evals = 10
opt_val, opt_pt, history = maximise_function(hartmann6_3, config.domain, max_num_evals,
                                           config=config)
print(opt_pt, opt_val)
```
