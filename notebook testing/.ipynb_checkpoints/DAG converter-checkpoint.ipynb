```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.dagcircuit import DAGCircuit
from qiskit.converters import circuit_to_dag
from qiskit.visualization import dag_drawer
from qiskit.circuit import Parameter
import networkx as nx
from typing import List
import numpy as np
%matplotlib inline
```

```python
q = QuantumRegister(3, 'q')
#c = ClassicalRegister(4, 'c')
theta = Parameter('Θ')
circ = QuantumCircuit(q)
circ.h(q[1])
circ.cx(q[0], q[1])
circ.ry(theta, q[0])
circ.rx(0.5, q[1])
circ.cry(theta, q[1],q[2])
#circ.measure(q,c)

dag = circuit_to_dag(circ)
nx_dag = dag.to_networkx()
dag_drawer(dag)
```

```python
in_nodes = [0]*dag.num_qubits()
out_nodes = [0]*dag.num_qubits()

for node in nx_dag.nodes():
    if node.type in ['in','out']:
        print(node.type, node.wire.index)
        
        if node.type == 'in':
            in_nodes[node.wire.index] = node
        if node.type == 'out':
            out_nodes[node.wire.index] = node
        
    elif node.type == 'op':
        print(node.name, [qubit.index for qubit in node.qargs])
```

```python
def longest_simple_path_length(graph, source, target) -> int:
    longest_paths = []
    longest_path_length = 0
    all_paths = nx.all_simple_paths(graph, source=source, target=target)
    
    if len(list(all_paths)) == 0:
        return len(dag.longest_path())
    
    for path in all_paths:    
        if len(path) > longest_path_length:
            longest_path_length = len(path)
            longest_paths.clear()
            longest_paths.append(path)
        elif len(path) == longest_path_length:
            longest_paths.append(path)
    return longest_path_length
```

```python
def shortest_simple_path_length(graph, source, target) -> int:
    shortest_paths = []
    shortest_path_length = 10e9
    all_paths = nx.all_simple_paths(graph, source=source, target=target)
    
    if len(list(all_paths)) == 0:
        return len(dag.longest_path())
    
    for path in all_paths:
        #print(len(path))
        if len(path) < shortest_path_length:
            shortest_path_length = len(path)
            shortest_paths.clear()
            shortest_paths.append(path)
        elif len(path) == shortest_path_length:
            shortest_paths.append(path)
    return shortest_path_length
```

```python
def structural_path_lengths(dag, nx_dag, in_nodes, out_nodes):
    n = dag.num_qubits()
    
    op_nodes = dag.op_nodes()
    lengths_to_in_nodes = np.zeros((len(op_nodes),n,2))
    lengths_to_out_nodes = np.zeros((len(op_nodes),n,2))
    
    for i,op_node in enumerate(op_nodes):
        for j,in_node in enumerate(in_nodes):
            lengths_to_in_nodes[i,j,0] = longest_simple_path_length(nx_dag, in_node, op_node)
            lengths_to_in_nodes[i,j,1] = shortest_simple_path_length(nx_dag, in_node, op_node)
            #lengths_to_in_nodes[i,j,2] = random_walk_path_length(nx_dag, in_node, op_node)
        for k,out_node in enumerate(out_nodes):
            lengths_to_out_nodes[i,j,0] = longest_simple_path_length(nx_dag, op_node, out_node)
            lengths_to_out_nodes[i,j,1] = shortest_simple_path_length(nx_dag, op_node, out_node)
            #lengths_to_out_nodes[i,j,2] = random_walk_path_length(nx_dag, op_node, out_node)
    return lengths_to_in_nodes, lengths_to_out_nodes
```

```python
lengths_to_in_nodes, lengths_to_out_nodes = structural_path_lengths(dag, nx_dag, in_nodes, out_nodes)
```
