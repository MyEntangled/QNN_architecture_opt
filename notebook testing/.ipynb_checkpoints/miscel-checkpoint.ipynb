```python
import numpy as np
import time
```

```python
num_samples = 100000
fidelities = np.random.uniform(0,1,(1000,num_samples))
k = 10
```

```python
save_a = np.zeros(1000)
save_b = np.zeros(1000)
save_c = np.zeros(1000)
```

```python
start_time = time.time()
a = np.sort(fidelities)
save_a = a[:,k]
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()
b = np.partition(fidelities,k)[:,:k+1]
save_b = np.sort(b)[:,-1]
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()
for i in range(1000):
    temp = np.partition(fidelities[i],k)[:k+1]
    save_c[i] = np.sort(temp)[-1]
print("--- %s seconds ---" % (time.time() - start_time))
```
