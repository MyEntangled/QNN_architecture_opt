```python
def func(x):
    u = x.reshape(x.shape[0],-1)
    t = u.T @ u
    val = np.trace(t) - np.log(np.linalg.det(t))
    
    history.append((x,val))
    return val
```

```python
def get_one_hot(x):
    gate_choice = x.reshape((x.shape[0], -1)).argmax(axis=1)
    u = np.zeros((x.shape[0], x.shape[1]*x.shape[2]))
    u[np.arange(x.shape[0]), gate_choice] = 1
    u = u.reshape(x.shape)
    
    return u
```

```python
def remove_id_gate(x):
    
    id_gate_column = x[:,:,0] # shape (num_gates, num_qubits)
     
    idx = np.argwhere(np.all(id_gate_column == 0,  axis=1)) # idx of non-identity gates
    
    u = np.zeros_like(x)
    
    u[:len(idx), ...] = x[idx.ravel()] # gather non-identity gates, pad the remaining positions with zeros
    
    #idx = np.argwhere(np.any(id_gate_column == 1,  axis=1))
    #x = np.delete(x, idx, axis=0)
    
    print(idx)
    
    return u
```
